生のデータを112x112にresizeし，さらにtrain/ valid/ testのスプリットを行なってFIleListを作るプログラム

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import sys
%matplotlib inline
```

```python
folder_normal = 'raw/normal'
folder_abnormal = "raw/abnormal/"

for curDir, dirs, normal_files in os.walk(folder_normal, topdown=False):
    print("normal")
    print(normal_files)

    
for curDir, dirs, abnormal_files in os.walk(folder_abnormal, topdown=False):
    print("abnormal")
    print(abnormal_files)
    

print(len(normal_files), len(abnormal_files))
```

### サイズを112x112にresizeして, videos/originalに保存
(すでにある場合は不要)

```python
def resize_avi_file(src_folder, src_filenames, output_folder, cropSize):
    if not os.path.exists(output_folder):
            os.mkdir(output_folder)
            
    for filename in src_filenames:
        
        file_path = os.path.join(src_folder ,filename)
        
        if not os.path.exists(file_path):
            raise FileNotFoundError(filename)
        
        print("resizing", filename, "to", output_folder)
        
        cap = cv2.VideoCapture(os.path.join(src_folder ,filename))
        if not cap.isOpened():
            sys.exit()

        fourcc = cv2.VideoWriter_fourcc('M','J','P','G')
        
        output_movie = cv2.VideoWriter(os.path.join(output_folder, filename), fourcc, 30, cropSize)

        while True:
            ret, frame = cap.read()

            if not ret:
                break

            frame = cv2.resize(frame, cropSize, interpolation = cv2.INTER_CUBIC)
            output_movie.write(frame)

        output_movie.release()
```

```python
# resize raw videos and save to videos/original

cropSize = (112, 112)
output_folder = "videos/original"
os.makedirs(output_folder, exist_ok=True)

resize_avi_file(folder_normal, normal_files, output_folder, cropSize)
resize_avi_file(folder_abnormal, abnormal_files, output_folder, cropSize)
```

### FileListを作る部分

```python
df_normal = pd.DataFrame(normal_files, columns=["filename"])
df_normal['target'] = 0
df_normal["patient_id"] = df_normal["filename"].map(lambda x: x.split("_")[0])
df_unique_patient_id = df_normal["patient_id"].value_counts().rename_axis('patient_id').reset_index(name='counts')
df_normal_rand=pd.concat([df_unique_patient_id, pd.DataFrame( np.random.rand( len( df_unique_patient_id ) ), columns=['rand'] ) ], axis=1 )
df_normal = pd.merge(df_normal, df_normal_rand, on='patient_id')
df_normal.head()
```

```python
# train : valid : test = 7 : 1.5 : 1.5
df_normal.loc[df_normal["rand"] > 0.3, 'split'] = 'TRAIN'
df_normal.loc[(df_normal["rand"] <= 0.3) & (df_normal["rand"] > 0.15), 'split'] = 'VALID'
df_normal.loc[df_normal["rand"] <= 0.15, 'split'] = 'TEST'
df_normal["split"].value_counts()
```

```python
df_abnormal = pd.DataFrame(abnormal_files, columns=["filename"])
df_abnormal['target'] = 1
df_abnormal["patient_id"] = df_abnormal["filename"].map(lambda x: x.split("_")[0])
df_unique_patient_id = df_abnormal["patient_id"].value_counts().rename_axis('patient_id').reset_index(name='counts')
df_abnormal_rand=pd.concat([df_unique_patient_id, pd.DataFrame( np.random.rand( len( df_unique_patient_id ) ), columns=['rand'] ) ], axis=1 )
df_abnormal = pd.merge(df_abnormal, df_abnormal_rand, on='patient_id')
df_abnormal.head()
```

```python
df_abnormal.loc[df_abnormal["rand"] > 0.3, 'split'] = 'TRAIN'
df_abnormal.loc[(df_abnormal["rand"] <= 0.3) & (df_abnormal["rand"] > 0.15), 'split'] = 'VALID'
df_abnormal.loc[df_abnormal["rand"] <= 0.15, 'split'] = 'TEST'
df_abnormal["split"].value_counts()
```

```python
df = pd.concat([df_normal, df_abnormal], axis=0 )
df["split"].value_counts()
```
